```python
f = plt.figure(figsize=(10,9))
ax1 = f.add_subplot(311)
ax2 = f.add_subplot(312)
ax3 = f.add_subplot(313)

fnum=8
c1weight=c2weight=c3weight=np.empty([0,fnum])


for iii in range(1,7):
    npyname='ldacoef'+str(iii)+'.npy'
    weight=np.load(npyname)
    weight.shape
    weight=np.abs(weight)
    weight.shape

    c1weight=np.vstack((c1weight,weight[0,:]))
    c2weight=np.vstack((c1weight,weight[1,:]))
    c3weight=np.vstack((c1weight,weight[2,:]))

    ax1.plot(np.arange(fnum),weight[0,:],'bo')
    #ax1.ylimit([0,5])
    ax2.plot(np.arange(fnum),weight[1,:],'bo')
    ax3.plot(np.arange(fnum),weight[2,:],'bo')
    
c1mean=np.mean(c1weight,axis=0)
c2mean=np.mean(c2weight,axis=0)
c3mean=np.mean(c3weight,axis=0)
c1var=np.var(c1weight,axis=0)
c2var=np.var(c2weight,axis=0)
c3var=np.var(c3weight,axis=0)

x_ticks_labels=['emg0','emg1','emg2','emg3','emg4','emg5','emg6','emg7']
ax1.errorbar(np.arange(fnum),c1mean,yerr=c1var,elinewidth=3,capsize=10,label='Errorbar',ls='none')
ax1.legend()
ax1.set_title('Class1 Relax',fontsize=14)
ax1.set_xticks(np.arange(fnum))
ax1.set_xticklabels(x_ticks_labels, fontsize=9)

ax2.errorbar(np.arange(fnum),c2mean,yerr=c2var,elinewidth=3,capsize=10,label='Errorbar',ls='none')
ax2.legend()
ax2.set_title('Class2 Open',fontsize=14)
ax2.set_xticks(np.arange(fnum))
ax2.set_xticklabels(x_ticks_labels, fontsize=9)

ax3.errorbar(np.arange(fnum),c3mean,yerr=c3var,elinewidth=3,capsize=10,label='Errorbar',ls='none')
ax3.legend()
ax3.set_title('Class3 Close',fontsize=14)
ax3.set_xticks(np.arange(fnum))
ax3.set_xticklabels(x_ticks_labels, fontsize=9)

f.tight_layout() 
#f.suptitle('Main title',fontsize=18)

plt.show()
```
